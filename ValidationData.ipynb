```python
import os

def supprimer_images_et_labels_sans_correspondance(images_path, labels_path):
    """
    Supprime les images du dossier 'images_path' qui n'ont pas de fichier de label correspondant dans 'labels_path',
    et supprime les fichiers de label qui n'ont pas d'image correspondante. Ne supprime pas le fichier 'classes.txt'.
    """
    # Récupérer les noms de fichiers sans extension pour les images et les labels
    image_files = {os.path.splitext(file)[0] for file in os.listdir(images_path) if file.endswith(('.jpg', '.jpeg', '.png'))}
    label_files = {os.path.splitext(file)[0] for file in os.listdir(labels_path) if file.endswith('.txt')}

    # Trouver les images sans fichier de label correspondant
    images_sans_label = image_files - label_files
    # Trouver les labels sans image correspondante
    labels_sans_image = label_files - image_files

    # Supprimer les images sans label
    for image_name in images_sans_label:
        image_path = os.path.join(images_path, f"{image_name}.jpg")  # Chemin de l'image
        os.remove(image_path)  # Supprimer l'image
        print(f"Image supprimée : {image_path}")

    # Supprimer les labels sans image correspondante, en évitant le fichier 'classes.txt'
    for label_name in labels_sans_image:
        if label_name != 'classes':  # Ne pas supprimer le fichier 'classes.txt'
            label_path = os.path.join(labels_path, f"{label_name}.txt")  # Chemin du label
            os.remove(label_path)  # Supprimer le label
            print(f"Label supprimé : {label_path}")

# Exécuter la fonction pour supprimer les fichiers sans correspondance
supprimer_images_et_labels_sans_correspondance(images_path, labels_path)
print("Note: Ce script est un pseudo-code et n'exécute pas la suppression des fichiers dans cet environnement.")
```

```python
# Liste pour stocker les noms des fichiers non conformes
non_conforming_files = []

# Vérifier si chaque ligne d'un fichier commence par "0" ou "1"
def check_lines_start_with_0_or_1(filename):
    with open(filename, 'r') as file:
        for line in file:
            if not line.startswith(("0 ", "1 ")):  # Vérifie si la ligne commence par "0 " ou "1 "
                return False
    return True

# Itérer sur tous les fichiers dans le dossier de labels, en excluant "classes.txt"
for label_file in os.listdir(labels_path):
    if label_file == 'classes.txt':  # Ignorer 'classes.txt'
        continue
    
    file_path = os.path.join(labels_path, label_file)
    
    # Ignorer les dossiers, traiter uniquement les fichiers
    if os.path.isfile(file_path):
        # Vérifier si le fichier respecte la condition
        if not check_lines_start_with_0_or_1(file_path):
            non_conforming_files.append(label_file)

# Afficher les résultats
if non_conforming_files:
    print("Fichiers non conformes trouvés :")
    for file in non_conforming_files:
        print(file)
else:
    print("Tous les fichiers sont valides et commencent par 0 ou 1.")
```

```python
import os
import glob
from collections import Counter

# Compter les instances de chaque classe
class_counts = Counter()

# Lire tous les fichiers .txt (sauf le classes.txt)
for label_file in glob.glob(os.path.join(labels_path, '*.txt')):
    if os.path.basename(label_file) == 'classes.txt':
        continue
    with open(label_file, 'r') as file:
        for line in file:
            class_id = int(line.split()[0])
            class_counts[class_id] += 1

# Afficher les comptages pour chaque classe
print(class_counts)
```

```python
import os

# Define the paths
labels_dir = 'dataset/labels'

# Iterate through all files in the directory
for filename in os.listdir(labels_dir):
    if filename.endswith('.txt'):
        file_path = os.path.join(labels_dir, filename)
        
        # Read the content of the file
        with open(file_path, 'r') as file:
            lines = file.readlines()
        
        # Modify the lines
        modified_lines = []
        for line in lines:
            parts = line.strip().split()
            if parts[0] == '0':
                parts[0] = '15'
            elif parts[0] == '1':
                parts[0] = '16'
            modified_lines.append(' '.join(parts))
        
        # Write the modified content back to the file
        with open(file_path, 'w') as file:
            for line in modified_lines:
                file.write(f"{line}\n")

print("Labels updated successfully.")
```
